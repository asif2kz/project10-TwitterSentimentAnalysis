```python
from nltk.corpus import stopwords
stop=stopwords.words('english')
import string
punctuations=list(string.punctuation)
stop += punctuations
stop
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif(tag.startswith('V')):
        return wordnet.VERB
    elif(tag.startswith('N')):
        return wordnet.NOUN
    elif(tag.startswith('R')):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
## clean the data
def clean_words(words):
    output=[]
    for w in words:
        if w.lower() not in stop:
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output.append(clean_word.lower())
    return output
```

```python
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
from nltk.corpus import wordnet
documents=[(clean_words(docwords),cat) for docwords,cat in data]
```

```python
a=[]
b=[]
for i,j in documents:
    a.append(" ".join(i))
    b.append(j)
```

```python
## now apply count vectorizer on this cleaned data to get feature and corresponding counts
import re
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    return text
from sklearn.feature_extraction.text import CountVectorizer
count_vec= CountVectorizer(max_features=1000,preprocessor=preprocess_text)
k=count_vec.fit_transform(a)
```

```python
count_vec.get_feature_names()
```

```python
training_data=k.todense()
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(training_data,b)
```
